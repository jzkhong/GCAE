# Introduction

Traditional autoencoders are widely used for dimensionality reduction and data reconstruction by focusing primarily on minimizing reconstruction error. While effective in these aspects, they often fall short in producing lower-dimensional embeddings that are useful for interpretability and tasks such as clustering and anomaly detection.

In recent years, constrained autoencoders have been proposed to address this issue by introducing various constraints to produce more useful embeddings. However, most focus on Euclidean distance or non-distance-based constraints, leaving the incorporation of geodesic distance constraints largely unexplored. Geodesic distance can provide a more accurate representation of the intrinsic geometry of complex, non-linear data manifolds, potentially enhancing the interpretability and usefulness of embeddings for clustering and anomaly detection.

This notebook conducts an exploratory analysis comparing Principal Component Analysis (PCA), Isomap, t-distributed Stochastic Neighbor Embedding (t-SNE), and traditional autoencoders in dimensionality reduction and producing useful embeddings for clustering. The analysis is performed on simulated datasets (S-Curve and Swiss Roll) and the MNIST digit image dataset. The aim is to demonstrate PCA's limitations with non-linear data, Isomap and t-SNE's effectiveness in producing 2D embeddings suitable for clustering, and traditional autoencoders' focus on data reconstruction, which does not always ensure useful embeddings. 

By highlighting these differences, this analysis underscores the potential benefits of incorporating geodesic distance constraints into autoencoders, paving the way for future research in developing more powerful and interpretable embedding methods for complex data.

Principal Component Analysis (PCA), Isomap, and t-Distributed Stochastic Neighbor Embedding (t-SNE) were applied to both the S-Curve and Swiss Roll datasets to obtain two-dimensional representations. The scatterplots of the first two principal components obtained from PCA display significant overlap among different clusters, failing to effectively reveal the underlying structure of the data. In contrast, Isomap embeddings successfully unfold the datasets into meaningful 2D representations, preserving the manifold's structure and capturing the non-linear relationships accurately. t-SNE embeddings, while providing clear separation of clusters and reflecting the intrinsic structure, do not always accurately preserve the actual geodesic distances between data points or the overall geometry of the underlying three-dimensional data. Instead, t-SNE focuses on maintaining the local structure and cluster separation.

These results demonstrates that PCA struggles with non-linear datasets, while Isomap excels in preserving the manifold's geometry, and t-SNE effectively reveals cluster structures but lacks accurate distance preservation.

### Dimensionality Reduction using Autoencoder

To explore the dimensionality reduction and data reconstruction capabilities of autoencoders, autoencoders were applied to the S-Curve and Swiss Roll datasets. Two separate autoencoders were built and trained independently on the S-Curve and Swiss Roll datasets. Both autoencoders leverage similar architecture:

- **Encoder**: The encoder had three dense layers. The first layer had 128 neurons with ReLU activation, the second layer had 64 neurons with ReLU activation, and the final layer reduced the dimensionality to the two-dimensional latent space with ReLU activation.
- **Decoder**: The decoder mirrored the encoder's structure. It began with a layer of 64 neurons with ReLU activation, followed by 128 neurons with ReLU activation, and the final layer reconstructed the input three-dimensional dataset.

```python
# Build and train the autoencoder for the S-Curve and Swiss Roll datasets
s_curve_autoencoder = get_autoencoder(3, 2, 's_curve')
s_curve_history = train_autoencoder(x[0], s_curve_autoencoder, epochs=100, batch_size=16, validation_split=0.1)
swiss_roll_autoencoder = get_autoencoder(3, 2, 'swiss_roll')
swiss_roll_history = train_autoencoder(x[1], swiss_roll_autoencoder, epochs=100, batch_size=16, validation_split=0.1)
# Reconstructs the S-Curve and Swiss Roll dataset
rc_s_curve = s_curve_autoencoder(x[0]).numpy()
rc_swiss_roll = swiss_roll_autoencoder(x[1]).numpy()
plot_3d_datasets((rc_s_curve, rc_swiss_roll), y, 'Reconstructed Data',
                 txt='Figure 3: 3D scatter plots of the reconstructed S-Curve (left) and Swiss Roll (right) datasets.')
```

```python
# Examine the latent encodings
s_curve_encodings = compute_latent_encodings(x[0], s_curve_autoencoder, 's_curve')
swiss_encodings = compute_latent_encodings(x[1], swiss_roll_autoencoder, 'swiss_roll')
visualise_latent_encodings(s_curve_encodings, y[0], 'S-Curve', swiss_encodings, y[1], 'Swiss Roll')
```

**Results on S-Curve and Swiss Roll Datasets**

The reconstructions from the autoencoder indicate that it effectively captured the overall shape and structure of both the S-Curve and Swiss Roll datasets. The scatter plots of the reconstructed data (Figure 3) show that the reconstructed data points have shapes and distributions that align closely with the original manifold, indicating successful dimensionality reduction and data reconstruction.

In terms of the two-dimensional latent embeddings, the results were mixed. The latent embeddings for both datasets (Figure 4) display a clear separation between clusters, indicating that the autoencoders captured the broad, non-linear structure of the data. This suggests that the autoencoders were effective in reducing the dimensionality while maintaining the overall manifold structure. However, the latent embeddings did not fully preserve the local and global geometrical relationships present in the original datasets. The embeddings made it difficult to interpret the exact relationships and distances between different clusters. This limitation implies that while the autoencoders are useful for reconstructing data and capturing general structures, the latent spaces they produce may not be optimal for data analysis that requires precise interpretability and spatial relationships.

## MNIST

The MNIST dataset is a well-known benchmark in the machine learning community, consisting of 60,000 training images and 10,000 testing images of handwritten digits, ranging from 0 to 9. Each image is a 28x28 pixel grayscale image, making it a high-dimensional dataset (784 dimensions). This dataset is ideal for testing dimensionality reduction techniques due to its complexity and the clear, categorical nature of its labels.

Before applying the dimensionality reduction techniques, the following preprocessing steps were performed:

1. Normalisation: Each pixel value in the MNIST images was normalised to lie between 0 and 1. This is achieved by dividing each pixel value by 255, the maximum pixel value in the dataset.
2. Flattening: The 28x28 pixel images were flattened into 784-dimensional vectors. This transformation is necessary for PCA, Isomap, and t-SNE, which require input data to be in an array format rather than a 2D image structure.
3. Standardisation: The flattened dataset was standardised to have a mean of 0 and a standard deviation of 1 before performing PCA. Standardisation helps in centering the data and scaling it to unit variance, which is important for PCA to work effectively.

```python
# Load the preprocessed mnist dataset and inspect 10 sample images
x_train, x_test, flat_x_train, flat_x_test, y_train, y_test, unique_labels = load_preprocessed_mnist()
visualise_sample_images(x_train, y_train)
```

### Dimensionality Reduction with PCA, Isomap and t-SNE

Applying PCA, Isomap, and t-SNE to the preprocessed MNIST dataset provides valuable insights into the capabilities and limitations of each dimensionality reduction technique when handling high-dimensional, real-world data:

- **PCA**: The scatter plot of the first two principal components from PCA revealed significant overlap among different digit classes, indicating that the clusters were not well-separated. This highlights PCA's limitation in capturing the complex, non-linear structures inherent in the MNIST data. PCA's reliance on linear assumptions is insufficient for datasets with intricate, non-linear relationships, resulting in poor performance in distinguishing the digit classes.

- **Isomap**: The embeddings from Isomap offered better separation between digit classes compared to PCA. The clusters were more distinct, reflecting Isomap's ability to preserve the global structure of the data's manifold. However, there was still some overlap between classes, particularly among the digits 2, 3, 5, and 8, where Isomap struggled to separate them into distinct clusters. While Isomap captures more meaningful relationships than PCA, it is not entirely effective in distinguishing all digit classes.

- **t-SNE**: The embeddings from t-SNE produced clear and distinct clusters for each digit class, with significantly better separation compared to PCA and Isomap. This makes t-SNE highly suitable for visualisation and exploratory data analysis. However, t-SNE focuses on preserving local structure, which means it does not accurately preserve global distances and relationships. Despite this, its ability to create well-separated clusters makes it a powerful tool for visualising high-dimensional data.

Note: Both Isomap and t-SNE can be computationally intensive and time-consuming, especially on large datasets. Therefore, only the first 30,000 training images were used to fit these algorithms instead of the full training dataset to manage computational resources effectively.

### CNN Autoencoder

An autoencoder was trained on the normalised MNIST dataset to compress the high-dimensional images into a two-dimensional latent space and then reconstruct the original images. The autoencoder architecture used a convolutional neural network (CNN) structure, which is particularly well-suited for image data. The CNN autoencoder architecture was as follows:

- **Encoder**: The encoder included two convolutional layers with ReLU activation, followed by two dense layers to reduce the dimensionality to the target latent space. The first convolutional layer had 32 filters, and the second had 64 filters, both with a kernel size of 3x3, strides of 2, and padding set to 'same'. After flattening, the first dense layer had 64 units, and the final dense layer compressed the data to the desired encoded dimension.
- **Decoder**: The decoder mirrored the encoder's structure, starting with two dense layers to expand the encoded representation. The first dense layer had 64 units, and the second had 3136 units (7x7x64), followed by two transposed convolutional layers to reconstruct the original image dimensions. The transposed convolutional layers used 32 filters for the first layer and the same number of channels as the input for the final layer, with ReLU and sigmoid activations respectively.

**Results of the CNN Autoencoder**

- Reconstructed Images: The reconstructed images from the CNN autoencoder closely resembled the original digits (as shown in Figure 7), demonstrating the model's effectiveness in learning and reproducing the high-dimensional data. This underscores the autoencoder's strength in capturing and reconstructing the overall structure and details of the MNIST images.

- Latent Embeddings: The two-dimensional latent embeddings (Figure 8) produced by the CNN autoencoder struggle to demonstrate clear separation between most digit classes. While there is some clustering, particularly for the digit '1', other digit classes are significantly overlapped and mixed. This suggests that while the autoencoder is effective for reconstruction, its embeddings are not optimized for tasks requiring clear class separation and interpretability.

# Conclusion

This exploratory analysis evaluated the performance of PCA, Isomap, t-SNE, and autoencoders on both simulated (S-Curve and Swiss Roll) and real-world (MNIST) datasets. Each technique was assessed for its ability to capture intrinsic data structures and produce meaningful lower-dimensional embeddings.

PCA, while computationally efficient, struggled with non-linear datasets, resulting in significant overlap in the embeddings. Isomap provided better cluster separation than PCA by preserving the global data structure, though it still exhibited some overlap in the MNIST dataset. t-SNE excelled in revealing clear and distinct clusters, making it ideal for visualisation, but it did not preserve global distances accurately. Autoencoders showed strong data reconstruction capabilities but produced less interpretable embeddings compared to Isomap and t-SNE, focusing on minimising reconstruction error rather than optimising for clustering tasks.

These findings highlight the importance of selecting the appropriate dimensionality reduction technique based on the specific requirements of the task. While PCA offers a good baseline and computational efficiency, Isomap and t-SNE perform better on non-linear datasets, with t-SNE being particularly effective for visualising complex data structures. Future research should explore integrating geodesic distance constraints into autoencoders to enhance the interpretability and practical application of embeddings. Such advancements could lead to more powerful and versatile dimensionality reduction techniques, well-suited for data visualisation, clustering, and anomaly detection in high-dimensional, non-linear datasets while still capable of accurate data reconstruction.
